```python
# Importing the csv file containing the data for R1 project
import pandas as pd
import numpy as np
Laliga=pd.read_csv('C:/PGP-AIML/R1_Project/Laliga.csv')
print (Laliga.head (2))
type (Laliga)
```

```python
# Ques 1: Read the data set and replace dashes with 0 for arithmetic operations
New_Header=Laliga.iloc[0]
Laliga_1=Laliga[1:]
Laliga_1.columns=New_Header
Laliga_1=Laliga_1.replace('-',0)
print (Laliga_1.head(10))
```

```python
# Ques 2: Define the Debut 1930-1980
Laliga_1['Debut_Year']=Laliga_1['Debut'].str[:4]
Laliga_1=Laliga_1.astype({'Debut_Year':'int64'})
Laliga_1.dtypes
Laliga_2=Laliga_1[(Laliga_1['Debut_Year'] >= 1930) & (Laliga_1['Debut_Year'] <= 1980) ]
print (Laliga_2)
```

```python
# Ques 3: Top 5 teams in points
Laliga_1['Points']=Laliga_1['Points'].apply(pd.to_numeric)
Laliga_1.dtypes
print (Laliga_1.nlargest(5,'Points'))
```

```python
# Ques 4: Alternate approach
Goal_diff_count = lambda a,b : a-b
Laliga_1[['GoalsFor','GoalsAgainst']]=Laliga_1[['GoalsFor','GoalsAgainst']].apply(pd.to_numeric)
Laliga_1['Goal_Difference']=Goal_diff_count(Laliga_1['GoalsFor'],Laliga_1['GoalsAgainst'])
print (Laliga_1)
print ('Team with max goal difference')
print ('------------------------------')
print (Laliga_1[Laliga_1.Goal_Difference == Laliga_1.Goal_Difference.max()])
print ('Team with min goal difference')
print ('------------------------------')
print (Laliga_1[Laliga_1.Goal_Difference == Laliga_1.Goal_Difference.min()])
Laliga_1.drop ('Goal_Difference',axis=1,inplace=True) 
```

```python
# Ques 5 : Winning percent
Laliga_1[['GamesWon','GamesPlayed']]=Laliga_1[['GamesWon','GamesPlayed']].apply(pd.to_numeric)
Laliga_1['Winning Percent']=(Laliga_1['GamesWon']/Laliga_1['GamesPlayed'])*100
Laliga_1['Winning Percent'].fillna(0, inplace=True)
print (Laliga_1.tail(5))
print ('Top 5 Teams with highest winning percentage')
print ('-------------------------------------------')
print (Laliga_1.nlargest(5,'Winning Percent'))
```

```python
# Ques 6 Group by Best Position with sum of points
Laliga_1[['BestPosition','Points']]=Laliga_1[['BestPosition','Points']].apply(pd.to_numeric)
Laliga_Q6=Laliga_1.groupby ('BestPosition', as_index=False)['Points'].sum()
print (Laliga_Q6)
```
