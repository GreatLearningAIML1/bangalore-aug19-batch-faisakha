```
bac_df['labels']=bac_df['gender']+bac_df['age'].astype(str)+bac_df['topic']+bac_df['sign']
bac_df.head(2)
```

```
bac_df1=bac_df.drop(['id','gender','age','topic','sign','date'],axis=1)
bac_df1.head(2)
```

```
#Q4: Separating Features and Labels, train/test split 
x=bac_df1['text']
y=bac_df1['labels']
print(x.head(2))
print(y.head(2))
```

```
from sklearn.model_selection import train_test_split
```

```
# split X and y into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=10)
```

```
##Training data
print(x_train.shape)
print(y_train.shape)
```

```
#Test data
print(x_test.shape)
print(y_test.shape)
```

```
#Q5 --> Vectorize the features
#Feed input data to CountVectorizer after including ngram
cvect = CountVectorizer(stop_words='english',lowercase=True,ngram_range=(1,2))
cvect.fit(x_train)

#Check the vocablury size
len(cvect.vocabulary_)
```

```
#Convert Training/Test data Count Vectors
x_train_ct=cvect.transform(x_train)
x_test_ct=cvect.transform(x_test)
```

```
#Size of Document Term Matrix
print(x_train_ct.shape)
print(x_test_ct.shape)
```

```
#Q7 --> Transform the labels
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()

```

```
y_train1=mlb.fit_transform(y_train)
y_train1
```

```
y_test1=mlb.fit_transform(y_test)
y_test1
```

```
#Q8 - choosing a classifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
```

```
clf=LogisticRegression(solver='lbfgs')
clf=OneVsRestClassifier (clf)
```

```
clf.fit(x_train_ct, y_train1)
```

```
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
from sklearn.metrics import f1_score,classification_report, confusion_matrix,accuracy_score
```

```
#Q-9
y_predict = clf.predict(x_test_ct)
print(metrics.accuracy_score(y_test1, y_predict))
print(metrics.confusion_matrix(y_test1, y_predict))
print(metrics.classification_report(y_test1, y_predict))
```
